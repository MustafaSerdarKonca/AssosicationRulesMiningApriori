```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

def find_association_rules(dataset, transaction_col, min_support, metric='lift', min_threshold=0.5):
    # Convert the dataset into a list of transactions
    transactions = []
    for i in dataset[transaction_col].unique():
        list_trans = list(set(dataset[dataset[transaction_col] == i]['Item']))
        if len(list_trans) > 0:
            transactions.append(list_trans)
    
    # Encode the transaction data into a one-hot encoded dataframe
    trans_encoding = TransactionEncoder()
    onehot = trans_encoding.fit_transform(transactions)
    df = pd.DataFrame(onehot, columns=trans_encoding.columns_)
    
    # Use Apriori algorithm to find frequent itemsets
    frequent_set = apriori(df, min_support=min_support, use_colnames=True)
    
    # Use association rules to generate rules with minimum threshold
    rules = association_rules(frequent_set, metric=metric, min_threshold=min_threshold)
    
    for index, row in rules.iterrows():
        antecedents = ', '.join(list(row['antecedents']))
        consequents = ', '.join(list(row['consequents']))
        support = round(row['support'], 3)
        confidence = round(row['confidence'], 3)
        lift = round(row['lift'], 3)
        print(f"{antecedents} --> {consequents} (support={support}, confidence={confidence}, lift={lift})")
        
        # create empty dictionaries for antecedents and consequents
    antecedent_dict = {}
    consequent_dict = {}

    # iterate through the dataset and store the antecedents and consequents for each item
    for index, row in rules.iterrows():
        antecedents = list(row['antecedents'])
        consequents = list(row['consequents'])

        for item in antecedents:
            if item not in antecedent_dict:
                antecedent_dict[item] = []
            antecedent_dict[item].append(', '.join(consequents))

        for item in consequents:
            if item not in consequent_dict:
                consequent_dict[item] = []
            consequent_dict[item].append(', '.join(antecedents))
    print('\n---------------------------------------------------------------------------------------------------\n')
    # print the associations in the desired format
    for item, consequents in consequent_dict.items():
        antecedents = ', '.join(sorted(set(antecedent_dict[item])))
        print(f"{item} --> {antecedents}")
```

```python
find_association_rules(df, 'Transaction', 0.05, metric='lift', min_threshold=1.0)
```
